## 1. Get title from html

```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = "http://www.nikkei.com/"
```

```python
# request row html text
html = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
print(soup)
```

```python
# get title element
title_tag = soup.title
# get only the spring from the element
title = title_tag.string
```

```python
print(title_tag)
print(title)
```

## 2. Get NikkeiHeikin

```python
url3 = "http://www.nikkei.com/markets/kabu/"
html3 = urllib.request.urlopen(url3)
soup = BeautifulSoup(html3, "lxml") # for class attribute
print(html3)
```

```python
span = soup.find_all("span")
print(span)
```

```python
nikkei_heikin = ""
```

```python
for tag in span:
    try:
        string_ = tag.get("class").pop(0)
        #print("good1")
        if string_ in "mkc-stock_prices":
            nikkei_heikin = tag.string
            #print("good")
            break
    except:
        #print("bad")
        pass
```

```python
print(nikkei_heikin)
```
